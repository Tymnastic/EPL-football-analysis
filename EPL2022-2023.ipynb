```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the epl_clubs_info_2022-23.csv file into df1 DataFrame
df1 = pd.read_csv("C:/Users/ONILEIMO TIMILEHIN/Downloads/epl_clubs_info_2022-23.csv")
df1.head()
```

```python
# Read the epl_results_2022-23.csv file into df DataFrame
df = pd.read_csv("C:/Users/ONILEIMO TIMILEHIN/Downloads/epl_results_2022-23.csv")
df.head()
```

```python
# Get basic statistics of the df DataFrame
df.describe()
```

```python
# Add 'Match_id' column with match IDs to the df DataFrame
df.insert(1, 'Match_id', ['M:{0:03d}'.format(i+1) for i in range(len(df))])

# Drop the 'Div' column from the df DataFrame
df.drop(columns=['Div'], inplace=True)
df.head()
```

```python
# Check for null values in the df DataFrame
df.isnull().sum()
```

```python
# Display information about the df DataFrame
df.info()
```

```python
import datetime

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Convert the 'Time' column to datetime format and format it as '%I:%M %p'
df['Time'] = pd.to_datetime(df['Time'], format='%b:%M').dt.strftime('%I:%M %p')

df.info()
df.head()
df
```

```python
# Rename columns in the df DataFrame
df.rename(columns={'HF': 'Home_fouls', 'AF': 'Away_fouls', 'HY': 'Home_offside', 'AY': 'Away_offside'})

# Concatenate the 'HomeTeam' and 'AwayTeam' columns into df_team Series
df_team = pd.concat([df['HomeTeam'], df['AwayTeam']])

df_team
```

```python
# Add a new column 'team_goal' with the sum of 'FTHG' and 'FTAG' columns
df['team_goal'] = df['FTHG'] + df['FTAG']
df['team_goal']
```

```python
# Group the DataFrame by 'HomeTeam' and sum the goals scored at home
homegoal = df.groupby('HomeTeam').sum()
homegoal
```

```python
# Group the DataFrame by 'AwayTeam' and sum the goals scored away
awaygoal = df.groupby('AwayTeam').sum()
awaygoal
```

```python
# Calculate total goals by adding home goals and away goals
goals = homegoal.FTHG + awaygoal.FTAG
goals
```

```python
# Create a DataFrame 'total_goals' with 'Teams' and 'TotalGoals' columns, sorted by total goals in descending order
total_goals = goals.to_frame().sort_values(by=0, ascending=False).reset_index().rename(columns={'HomeTeam': 'Teams', 0: 'TotalGoals'})
total_goals
```

```python
import seaborn as sns
# Plot a bar chart showing the total goals scored by each team
plt.figure(figsize=(10, 5))
plt
plt.xlabel('Total Goals')
plt.ylabel('Teams')
plt.title('TOTAL GOALS SCORED BY TEAM')
figx = sns.barplot(data=total_goals, x='TotalGoals', y='Teams')
ax = figx.axes
for p in ax.patches:
    ax.text(x=p.get_width(), y=p.get_y() + p.get_height() / 2, s=int(p.get_width()), ha="left", va="center")

```

```python
# Calculate goals against by adding away goals and home goals against
GoalsAgainst = homegoal.FTAG + awaygoal.FTHG
Goals_Against = GoalsAgainst.to_frame().sort_values(by=0, ascending=True).reset_index().rename(columns={'HomeTeam': 'Team', 0: 'GoalsAgainst'})

plt.figure(figsize=(10, 5))
plt.xlabel('Goals Against')
plt.ylabel('Team')
plt.title('TOTAL GOALS SCORED AGAINST')
# Create a bar plot showing the total goals scored against each team
fig2 = sns.barplot(data=Goals_Against, x='GoalsAgainst', y='Team')
ax = fig2.axes

# Add data value labels to each bar
for p in ax.patches:
    ax.text(x=p.get_width(), y=p.get_y() + p.get_height() / 2, s=int(p.get_width()), ha="left", va="center")
```

```python
# Count the number of matches officiated by each referee
df['Referee'].value_counts()

# Plot a bar chart showing the number of matches officiated by each referee
df['Referee'].value_counts().plot(kind='bar', color='blue', figsize=(10, 6))
plt.xlabel('Referee')
plt.ylabel('Matches')
plt.title('Number of Officiated Matches By Referees')
plt.show()
```

```python
# Calculate the total corners for each team by summing home corners and away corners
homecorners = df.groupby('HomeTeam').sum()
awaycorners = df.groupby('AwayTeam').sum()

# Calculate the total corners for each team and create a DataFrame 'col'
col = (homecorners['HC'] + awaycorners['AC']).to_frame().sort_values(by=0, ascending=False).reset_index().rename(columns={'HomeTeam': 'Teams', 0: 'Corners'})

# Create a color palette for the bar plot
palette = sns.color_palette('cubehelix', n_colors=len(col['Teams'].unique()))

# Plot a bar chart showing the total corners for each team
fig3=sns.barplot(data=col, x='Corners', y='Teams', palette=palette)
plt.figure(figsize=(12,8 ))
ax=fig3.axes
# Add data value labels to each bar
for p in ax.patches:
    ax.text(x=p.get_width(), y=p.get_y() + p.get_height() / 2, s=int(p.get_width()), ha="left", va="center")
```

```python
# Calculate the total number of offside incidents for each team
F = df.groupby('HomeTeam')['HY'].sum()
offside = df.groupby('HomeTeam')['HY'].sum() + df.groupby('AwayTeam')['AY'].sum()
offsides = offside.to_frame().sort_values(by=0, ascending=False).reset_index().rename(columns={'HomeTeam': 'Team', 0: 'Offsides'})

# Plot a bar chart showing the total number of offside incidents for each team
offsides.plot.bar(x='Team', y='Offsides', color=['r'], figsize=(10, 8))
plt.xlabel('Team')
plt.ylabel('No of Offside')
plt.title('Total offsides')
plt.show()
```

```python
# match wins
def match_outcome(row):
    if row['FTR'] == 'H':
        return row['HomeTeam'] + ' Win'
    elif row['FTR'] == 'A':
        return row['AwayTeam'] + ' Win'
    else:
        return 'Draw'

# Apply the match_outcome function to each row in the DataFrame to create a new column 'Outcome'
df['Outcome'] = df.apply(match_outcome, axis=1)

teamwins=df[df['Outcome'] !='Draw'].Outcome.value_counts().to_frame().reset_index().rename(columns={'index':'Team','Outcome':'Wins'})
# Plot a bar chart to show the distribution of match outcomes
plt.figure(figsize=(8, 6))
fig4=sns.barplot(data=teamwins,x= 'Wins', y='Team')
ax=fig4.axes
# Add data value labels to each bar
for p in ax.patches:
    ax.text(x=p.get_width(), y=p.get_y() + p.get_height() / 2, s=int(p.get_width()), ha="left", va="center")

plt.axis('equal')
plt.title('Match Outcomes')
plt.show()

```

```python
# Calculate the average number of goals scored per match for each team
home_goals_avg = df.groupby('HomeTeam')['FTHG'].mean()
away_goals_avg = df.groupby('AwayTeam')['FTAG'].mean()
total_goals_avg = (home_goals_avg + away_goals_avg) / 2
# Sort the average goals scored in descending order and create a DataFrame 'avg_goals'
avg_goals = total_goals_avg.to_frame().sort_values(by=0, ascending=False).reset_index().rename(columns={'HomeTeam': 'Teams', 0: 'AverageGoalsPerMatch'})
avg_goals
```

```python
# Plot a horizontal bar chart to show the average number of goals scored per match by each team
import seaborn as sns
plt.figure(figsize=(10, 6))
fig5=sns.barplot(data=avg_goals, x='AverageGoalsPerMatch', y='Teams')
ax=fig5.axes
# Add data value labels to each bar
for p in ax.patches:
    ax.text(x=p.get_width(), y=p.get_y() + p.get_height() / 2, s=f'{p.get_width():.2f}', ha="left", va="center")

plt.xlabel('Average Goals')
plt.ylabel('Teams')

plt.title('Average Goals Scored per Match by Team')
plt.show()


```
